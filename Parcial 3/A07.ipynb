```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score, roc_auc_score
```

```python
data = pd.read_csv("Advertising.csv")
data
```

```python
X = data.drop(columns=['sales'])
y = data['sales']
```

```python
model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
r2
```

En el siguiente modelo hacemos una regresión lineal simple podemos ver la variablilidad del modelo por lo tanto, aqui todavía no dividimos los datos en un train, test por lo que todavía no entrenamos nuestro modelo.

```python
#Usar un Decision Tree Regressor y ajustar el dataset
model_dt = DecisionTreeRegressor(random_state=42)
model_dt.fit(X, y)
y_pred_dt = model_dt.predict(X)
r2_dt = r2_score(y, y_pred_dt)
r2_dt
```

En el siguiente modelo todavía no ajustamos nuestro modelo con un test y train por lo tanto el modelo lo que hace es dividir y dividir nuestro modelo hasta que llega a un ajuste perfecto 1 , sin embargo, esto no es lo optimo debido a que, el modelo tiene un sobreajuste.

### Modelo con train y test 

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
r2_split = r2_score(y_test, y_pred)
r2_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
model_dt = DecisionTreeRegressor(random_state=42)
model_dt.fit(x_train, y_train)
y_pred_dt = model_dt.predict(x_test)
r2_split_dt = r2_score(y_test, y_pred_dt)
r2_split_dt
```

En los siguientes modelos podemos ver que en la regresión lineal simple sin el train test nos dio mejores resultados esto se debe en gran parte porque tiene menos varianza y al momento de entrenar nuestro modelo por lo que nos muestra  un menor valor pero se podría decir que este está más validado que el otro. Por otra parte, si nos ponemos a analizar el Decisión tree vemos muchos mejores resultados sin contar el Decisión tree sin el train test que sabemos que siempre nos va a dar uno por ese sobreajuste. Lo que hace el decisión tree es crear un espacio en el cual se adapta mejor a los datos. 

```python
# Repetir el proceso con PolynomialFeatures(degree = 2)
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(X)

model_poly = LinearRegression()
model_poly.fit(X_poly, y)

y_pred_poly = model_poly.predict(X_poly)

r2_poly = r2_score(y, y_pred_poly)
r2_poly
```

```python
model_dt = DecisionTreeRegressor(random_state=42)
model_dt.fit(X_poly, y)

y_pred_dt = model_dt.predict(X_poly)

r2_dt = r2_score(y, y_pred_dt)
r2_dt
```

El polinomial feauture que hace es que trabaja como un kernel en el cual, estas transformando tus variables a cierto grado. Esto afecta los parametros por lo tanto podemos ver una ,ejora en nuestra regresión lineal ya que aumenta la capacidad del modelo. Nuestro decisión tree sin train test es 1 por las mismas razones mencionadas anteriormente.

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)

X_poly_t = poly.fit_transform(x_train)

model.fit(X_poly_t, y_train)
y_pred = model.predict(poly.transform(x_test))
r2_split = r2_score(y_test, y_pred)
r2_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)

model_dt = DecisionTreeRegressor(random_state=42)
X_poly_D = poly.fit_transform(x_train)
model_dt.fit(X_poly_D, y_train)
y_pred_dt = model_dt.predict(poly.transform(x_test))
r2_split_dt = r2_score(y_test, y_pred_dt)
r2_split_dt
```

Finalmente podemos comparar nuestros 4 modelos despues de realizar un train test, definitavamente cuando aplicamos la transformación polynomial tenemos mejor resultados en nuestro r2 score sin embargo, podemos ver que el mejor modelo es la regresión logistica sin un train test, aunque al momento de aplicar el train test y el polinomial tenemos los resultados mucho más cercanos y una mejora en el ajuste de nuestro modelo. Junto a esto, podemos ver que el Decisión tree con la transformación poly tiene mejores resultados al momento. Esto se debe a que como aumentamos nuestra capacidad de modelo y nuestros parametros el modelo es mucho más certero
