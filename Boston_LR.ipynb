```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
df = pd.read_excel("Boston Housing Dataset 1978.xlsx")
data = df.dropna()
data.tail(5)
```

```python
data.describe()
```

```python
X = data[["CRIM","ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"]]
y = data["MEDV"]
```

```python
lr = LinearRegression()
lr.fit(X,y)

B0 = lr.intercept_
B1 = lr.coef_
B0,B1
```

```python
y_hat = lr.predict(X)
```

```python
X2 = sm.add_constant(X)
```

```python

ols = sm.OLS(y,X2)
results = ols.fit()
results.summary()
```

```python
for col in data.columns:
    if col == "MEDV":
        break
    min_d1 = data[col].min()
    data[col] - min_d1
    max_d1 = data[col].max()
    data[col] = data[col]/max_d1
data.describe()
```

```python
XD1 = sm.add_constant(data)
ols = sm.OLS(y,XD1)
results = ols.fit()
results.summary()
```
