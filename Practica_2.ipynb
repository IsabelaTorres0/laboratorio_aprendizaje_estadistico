## Aguas Frescas

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
df = pd.read_excel('Aguas_fresacas.xlsx')
df
```

```python
x = df.drop(columns=['cal'])
y = df['cal']
```

```python
ols = sm.OLS(y, sm.add_constant(x))
ols = ols.fit()
ols.summary()
```
