## Regresión polinomial

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Observa la siguiente gráfica.

```python
import matplotlib.pyplot as plt; import numpy as np
%matplotlib inline
np.random.seed(0)
x = np.linspace(-3, 3, 100); y = 1 + 4*x**2 + np.random.normal(0, 1, 100)
plt.figure(); plt.xlabel("x"); plt.ylabel("y")
plt.scatter(x, y); plt.grid()
```

Utiliza `sklearn` para realizar una regresión lineal simple ($\hat{y} = \hat{\beta_0} + \hat{\beta_1}x $)

```python
# Ajuste de regresión lineal simple
X = x.reshape(-1, 1)
lr = LinearRegression()
lr.fit(X,y)
y_hat = lr.predict(X)

y_hat
```

```python
#beta0 =
b0 = lr.intercept_
#beta 1
b1 = lr.coef_
b0,b1
```

Crea una gráfica donde se muestren las observaciones como una nube de puntos y la línea que ajustaste con regresión. Calcula el métrico $R^2$.

```python
plt.figure()
plt.scatter(x, y)
plt.plot(x, y_hat)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regresión lineal simple")
plt.grid()
plt.show()
```

```python
r2 = lr.score(X, y)
r2
```

Podemos ver que el ajuste no es correcto, pues parece que tiene un comportamiento cuadrático y estamos intentando ajustar una línea recta.

En lugar de hacer una regresión lineal simple, quizás podamos extender el modelo para que tenga varios factores dentro:

$$ Y = \beta_0 + \beta_1 X + \beta_2 X^2 + ... + \beta_p X^p + \epsilon$$

Ajusta un modelo cuadrático con `sklearn` ($\hat{y} = \hat{\beta_0} + \hat{\beta_1}x + \hat{\beta_1}x^2 $) y calcula su $R^2$.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
```

```python
X2 = np.hstack([
    X,
    X**2
])
lr.fit(X2,y)
#prediccion
y_pred2 = lr.predict(X2)

r_2 = r2_score(y, y_pred2)
# Betas 
#beta0_2
B0_2 = lr.intercept_
#beta1_2
B1_2 = lr.coef_

B0_2, B1_2
```

Crea una gráfica con las observaciones y la nueva curva que ajustaste.

```python
# Grado2
plt.figure()
plt.scatter(x, y)
plt.plot(x, y_pred2)
plt.title("Regresion con nueva curva de ajuste, grado 2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show
```

En si el ajuste de la curva es el grado 2 ya que, en los otros coeficcientes los cambios son minimos

Ajusta modelos de grados 3, 5, 7, 9, y 11. Calcula su $R^2$.

```python
#Ajuste de modelos
X3 = np.hstack([
    X, X**2, X**3
])

lr3=LinearRegression()
lr3.fit(X3,y)
#prediccion
y_pred3 = lr3.predict(X3)

r2_3 = r2_score(y, y_pred3)

# Betas 
#beta0_2
B0_3 = lr3.intercept_
#beta1_2
B1_3 = lr3.coef_

#B0_3, B1_3

X5 = np.hstack([
    X, X**2, X**4,X**5
])
lr5=LinearRegression()
lr5.fit(X5,y)
#prediccion
y_pred5 = lr5.predict(X5)

r2_5 = r2_score(y, y_pred5)

# Betas 
#beta0_5
B0_5 = lr5.intercept_
#beta1_5
B1_5 = lr5.coef_

#B0_5, B1_5

X7 = np.hstack([
    X, X**2, X**4,X**5,X**6, X**7
])

lr7=LinearRegression()
lr7.fit(X7,y)
#prediccion
y_pred7 = lr7.predict(X7)

r2_7 = r2_score(y, y_pred7)

# Betas 
#beta0_2
B0_7 = lr7.intercept_
#beta1_2
B1_7 = lr7.coef_

#B0_7, B1_7

X9 = np.hstack([
    X, X**2, X**4,X**5,X**6, X**7, X**8, X**9
])
lr9=LinearRegression()
lr9.fit(X9,y)
#prediccion
y_pred9 = lr9.predict(X9)

r2_9 = r2_score(y, y_pred9)

# Betas 
#beta0_2
B0_9 = lr9.intercept_
#beta1_2
B1_9 = lr9.coef_

#B0_9, B1_9, r2_9

X11 = np.hstack([
    X, X**2, X**4,X**5,X**6, X**7, X**8, X**9,X**10, X**11
])
lr11=LinearRegression()
lr11.fit(X11,y)
#prediccion
y_pred11 = lr11.predict(X11)

r2_11 = r2_score(y, y_pred11)

# Betas 
#beta0_2
B0_11 = lr11.intercept_
#beta1_2
B1_11 = lr11.coef_

#B0_11, B1_11
```

```python
#camparacion de las betas
print("Beta0 y beta 1 del grado 3")
print(B0_3, B1_3)
print()

print("Beta0 y beta 1 del grado 5")
print(B0_5, B1_5)
print()

print("Beta0 y beta 1 del grado 7")
print(B0_7, B1_7)
print()

print("Beta0 y beta 1 del grado 9")
print(B0_9, B1_9)
print()

print("Beta0 y beta 1 del grado 11")
print(B0_11, B1_11)
```

Recuerda que una de las metas de ajustar modelos es para poder interpolar y extrapolar. Los datos que tenemos mantienen $x \in [-3, 3]$. ¿Qué pasaría si quisiéramos saber el valor de $y$ cuando $x=5$?

Calcula la predicción $\hat{y}$ para todos los modelos creados (grados 2, 3, 5, 7, 9, y 11).

```python
# Valor para extrapolación
x_new = 5
x_new_array = np.array([[x_new]])

X3_n = np.hstack([x_new_array, x_new_array**2, x_new_array**3])
y_hat_3 = lr3.predict(X3_n)

X5_n = np.hstack([x_new_array, x_new_array**2, x_new_array**4, x_new_array**5])
y_hat_5 = lr5.predict(X5_n)

X7_n = np.hstack([x_new_array, x_new_array**2, x_new_array**4, x_new_array**5,
                    x_new_array**6, x_new_array**7])
y_hat_7 = lr7.predict(X7_n)

X9_n = np.hstack([x_new_array, x_new_array**2, x_new_array**4, x_new_array**5,
                    x_new_array**6, x_new_array**7, x_new_array**8, x_new_array**9])
y_hat_9 = lr9.predict(X9_n)

X11_n = np.hstack([x_new_array, x_new_array**2, x_new_array**4, x_new_array**5,
                     x_new_array**6, x_new_array**7, x_new_array**8, x_new_array**9,
                     x_new_array**10, x_new_array**11])
y_hat_11 = lr11.predict(X11_n)
```
